```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "name" : ["billy","Noah","mike","steve","michael"],
    "salary" : [450000,400000,350000,320000,250000],
    "joining_date" : ['2000-04-12','2003-04-27','2008-04-18','2009-04-18','2015-04-19'],
    "exit_date" : ['2022-03-18','2021-05-28','2019-07-28','2025-09-21','2024-10-29']
})
```

```python
df.info()
```

```python
df
```

```python
df['joining_date'] = pd.to_datetime(df['joining_date'])
df['exit_date'] = pd.to_datetime(df['exit_date'])
```

```python
df
```

```python
df['joining_year'] = df['joining_date'].dt.year
df['joining_month'] = df['joining_date'].dt.month
df['joining_day'] = df['joining_date'].dt.day


```

```python
df['days_since_join'] = (pd.Timestamp.today() - df['joining_date']).dt.days
```

```python
df['month_name'] = df['joining_date'].dt.month_name()
```

```python
df
```

```python
df = df.sort_values(by='joining_date', ascending=False)

```

```python
df.reset_index(drop=True)
```

```python
filtered = df[df['joining_date'] == '2000-04-12']
```

```python
joining_date = '2000-04-12'
end_date = '2025-09-28'
filtered_range = df[(df['joining_date'] >= joining_date) & (df['joining_date'] < end_date)]
```

```python
filtered_range
```

```python
start = '2005-01-01'
end   = '2010-01-01'

filtered_range = df[(df['joining_date'] >= start) & (df['joining_date'] < end)]
print(filtered_range)
```

```python
# every syntax should be separate 
df[df['joining_date'].dt.year == 2000] #by year
df[df['joining_date'].dt.month == 4] # By month
df[df['joining_date'].dt.day == 19] # by date
```

 ### For MONTHS AND YEAR WE USE DATE OFFSET 
 #### BUT WHEN WE NEED PARTICULAR DAY/TIME/ HOUR WE USE TIMEDELTA

```python
# Add 5 years (future date)
df['future_date'] = df['joining_date'] + pd.DateOffset(years=5)

# Subtract 2 years (past date)
df['past_date'] = df['joining_date'] - pd.DateOffset(years=2)
```

```python
df
```

```python
from datetime import timedelta

df['future_30days'] = df['joining_date'] +  timedelta(days = 30)
df['past_30days'] = df['joining_date'] - timedelta(days = 30)


```

```python
df['total_days'] = (df['exit_date'] - df['joining_date']).dt.days
```

```python
df['total_year'] = df['total_days']/365
```

```python
df['total_seconds'] = (df['exit_date'] - df['joining_date']).dt.total_seconds()
```
