```python
import pandas as pd

data = {
    'date': [
        '2024-01-01','2024-01-02','2024-01-03',
        '2024-01-04','2024-01-05','2024-01-06',
        '2024-01-07'
    ],
    'sales': [10, 20, 30, 40, 50, 60, 70]
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

```python
df
```

```python
df.resample('W').sum()
```

```python
df.resample('M').sum()
```

```python
import pandas as pd
import numpy as np

s = pd.Series([1, np.nan, np.nan, 4, np.nan])

print("Forward fill:\n", s.ffill())
print("Backward fill:\n", s.bfill())
```

```python
df.resample('M', label='left', closed='left').sum()
```

```python
df.resample('W-thu').sum()   # Weeks ending on Monday
```

```python
df.resample('M').agg({'sales':'sum'})
```
