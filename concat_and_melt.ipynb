```python
import pandas as pd
```

```python
df1 = pd.DataFrame ({
    'name':['bob','harley','harry','peter'],
    'age':[23,24,23,27]
})
df2 = pd.DataFrame ({
    'name':['hopper','butler','ricky','pattison'],
    'age':[25,24,23,29]
})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis = 0)
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
df2 = pd.DataFrame ({
    'name': ['james','suresh','ramesh'],
    'maths':[76,67,82],
    'science':[86,72,69]
})
```
