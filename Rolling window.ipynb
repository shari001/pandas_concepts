```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'sales': [100, 120, 130, 90, 80, 110, 150]
})
```

```python
df['sales_avg'] = df['sales'].rolling(window = 3).mean()
```

```python
df['sales_sum'] = df['sales'].rolling(window = 3).sum()
```

```python
df['rolling_avg'] = df['sales'].rolling(window = 3, min_periods=2).mean()
```
