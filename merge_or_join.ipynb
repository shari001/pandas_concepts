```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({"name": ["Bob","Harley","Nolan","Noah"],
                    "emp_id" : [101,102,103,104]
                  })
```

```python
df2 = pd.DataFrame({"emp_id" : [101,103,105],
                    "salary" : [20000,40000,50000]
})
```

```python
pd.merge(df1,df2, on = "emp_id", how = "inner")
```

```python
pd.merge(df1,df2, on = "emp_id", how = "left")
```

```python
pd.merge(df1,df2, on = "emp_id", how = "right")
```

```python
pd.merge(df1,df2,on = "emp_id", how = "outer")
```

```python
df1.set_index("emp_id").join(df2.set_index("emp_id"),how = "inner")
```

```python
df1.set_index("emp_id",inplace = True)
df2.set_index("emp_id",inplace = True)
```

```python
df1.join(df2,how = "inner")
```
