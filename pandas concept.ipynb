```python
import pandas as pd
import numpy as np
```

```python
data_emp    = {'name':['suresh','ramesh','alice','bob','mahesh','virat','sachin','gill','shubham','sonam','jagu','ishan','kisan','harry','sundhar','dulqar salman'],
               'age': [21,23,23,24,26,32,np.nan,45,32,np.nan,34,34,41,42,23,np.nan],
               'salary':[10000,30000,32000,20000,29000,45000,34000,10000,21000,10020,38000,23000,23900,21000,45000,30000]}
df = pd.DataFrame(data_emp)
```

```python
df['age'].fillna(25,inplace = True)
```

```python
df['age'].count()
```

```python
len(df['age'])
```

```python
df['department'] = [
    'HR','HR','Finance','Finance','IT','IT','Sales','Sales',
    'HR','Finance','IT','Sales','HR','Finance','IT','Sales'
]
```

```python
df
```

```python
df.loc[df['department'] == 'HR','salary'] = df['salary']  + 50000
```

```python
df
```

```python
df.groupby('department')['salary'].mean()
```

```python
df.groupby('age')['salary'].mean()
```

```python
df.groupby('age').agg({
    'salary':['min','max','mean'],
    'name': 'count'
})
```

```python
df['level'] = df['salary'].apply(lambda x: 'High' if x > 30000 else 'Low')
```

```python
df
```

```python
df['employee'] = df['age'].apply(lambda x : 'Experience' if x > 25 else 'Fresher')
```

```python
df
```

```python
df['bonus'] = df['salary'].apply(lambda x : x*0.1)
```

```python
df
```

```python
df['tax'] = df['bonus'].apply(lambda x : 0 if x <2500 
                                   else 0.1*x if x<(3000 and 4000)
                                    else  x*0.12)
```

```python
df
```

```python
df['age_group'] = df['age'].apply(lambda x: 'young' if x <30 
                                        else 'middle' if x <40
                                        else 'senior')
```

```python
df
```

```python
df['mask_name'] = df.apply(lambda row : row ['name'].upper() if row['age']>30
                             else  ['name'],axis = 1)
```
