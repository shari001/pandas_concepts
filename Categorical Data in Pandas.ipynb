```python
import pandas as pd
```

```python
sizes = pd.Series(['small','large','small','large','medium','large','medium']).astype('category')
```

```python
size = pd.Categorical(['small','large','small','large','medium','large','medium'],
                   categories = ['small','medium','large'],
                   ordered = True
)
print(sizes)
```

```python
sizes.cat.categories
```

```python
sizes.cat.codes
```

```python
sizes = sizes.cat.rename_categories(['Small','Medium','Large'])
print(sizes)
```

```python
sizes = sizes.cat.add_categories(['Extra'])
print(sizes)
```

```python
sizes = sizes.cat.remove_unused_categories()
print(sizes)
```

```python
sizes = sizes.cat.set_categories(['Tiny','Big','Huge'])
print("After set_categories:", sizes.cat.categories)

```

```python
sizes = sizes.cat.reorder_categories(['Tiny','Big','Huge'], ordered = True)
```

```python
print(sizes.sort_values()) 
```
